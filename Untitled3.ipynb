```python
from bs4 import BeautifulSoup
import urllib
from textblob import TextBlob
import matplotlib.pyplot as plt
link="https://bit.ly/2M9CrRL"
html_file =urllib.request.urlopen(link)
soup = BeautifulSoup(html_file,'html.parser')
```

```python
reviews= soup.find_all('p')
neutral=0
good=0
excellent=0
bad=0
worst=0
count=0
analysis=[neutral,good,excellent,bad,worst]
for i in reviews:
    count=count+1
    review=i.get_text()
    sentiment=TextBlob(review).sentiment.polarity
    if(sentiment==0):
        neutral=neutral+1
    if((sentiment<0) and(sentiment>=(-0.5))):
        bad=bad+1
    if(sentiment<(-0.5)):
        worst=worst+1
    if((sentiment>0) and (sentiment<=0.5)):
        good=good+1
    if(sentiment>0.5):
        neutral=neutral+1
plt.pie(analysis,labels=['Neutral','Good','Excellent','Bad','Worst'],shadow= True)
plt.title("Analysis for reviews/comments on above mentioned post")
plt.show()

    
```
