```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name=input(" Folder location ending with a backslash ")
File=input(" Enter file name such as abc.csv  ")
data = pd.read_csv(name+File)
print("  ")
print("International")
plt.plot(data.iloc[:,0],data.iloc[:,1])
plt.show()

print("Domestic")
plt.plot(data.iloc[:,0],data.iloc[:,2])
plt.show()

print("Total")
plt.plot(data.iloc[:,0],data.iloc[:,3])
plt.show()
```
