```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('https://query.data.world/s/bnnhbevc63n5xaq2ws7oa3oa3ovboe')
df.head()
```

```python
data = df.iloc[1:,[5,6]]
data.fillna(0.0)
data.head()
```

```python
growth=[]
for i in range(253):
    b=data.iloc[i,1]-data.iloc[i,0]
    growth.append(b)
data['Growth']= pd.Series(growth,  index=data.index)
data.head()
```

```python
arr=[]
for i in range(253):
    if data.iloc[i,2]>0:
        arr.append(1)
    else:
        arr.append(0)
data['Target Variable']= pd.Series(arr,  index=data.index)
data.head()
```

```python
data_hist= df.iloc[1:,4]
data_bar= df.iloc[1:,10]
data_pie= df.iloc[1:,[9,10,93,108,112,160]]
```

```python
data_hist.head()
```

```python
print("Pie Chart for establishment year")
yr1910=0
yr1920=0
yr1930=0
yr1940=0
yr1950=0
yr1960=0
yr1970=0
yr1980=0
yr1990=0
yr2000=0
yr2010=0


for i in range(253):
    if (float(data_hist.iloc[i])<=1920):
        yr1910+=1
    if (float(data_hist.iloc[i])<=1930) and (float(data_hist.iloc[i])>1920):
        yr1920+=1
    if (float(data_hist.iloc[i])<=1940)and (float(data_hist.iloc[i])>1930):
        yr1930+=1
    if (float(data_hist.iloc[i])<=1950)and (float(data_hist.iloc[i])>1940):
        yr1940+=1
    if (float(data_hist.iloc[i])<=1960)and (float(data_hist.iloc[i])>1950):
        yr1950+=1
    if (float(data_hist.iloc[i])<=1970)and (float(data_hist.iloc[i])>1960):
        yr1960+=1
    if (float(data_hist.iloc[i])<=1980)and (float(data_hist.iloc[i])>1970):
        yr1970+=1
    if (float(data_hist.iloc[i])<=1990)and (float(data_hist.iloc[i])>1980):
        yr1980+=1
    if (float(data_hist.iloc[i])<=2000)and (float(data_hist.iloc[i])>1990):
        yr1990+=1
    if (float(data_hist.iloc[i])<=2010)and (float(data_hist.iloc[i])>2000):
        yr2000+=1
    if (float(data_hist.iloc[i])<=2020)and (float(data_hist.iloc[i])>2010):
        yr2010+=1
values=[yr1910,yr1920,yr1930,yr1940,yr1950,yr1960,yr1970,yr1980,yr1990,yr2000,yr2010] 
index=['1910-1920','1920-1930','1930-1940','1940-1950','1950-1960','1960-1970','1970-1980','1980-1990','1990-2000','2000-2010','2010-till date']
plt.pie(values,labels=index,shadow= True)
plt.show()
```
